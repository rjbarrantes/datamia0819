```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

vehicles = pd.read_csv('vehicles.csv')
```

### Data Pre-processing (Handling Numerical variables)

```python
nums = vehicles[['Engine Displacement', 'Fuel Barrels/Year', 'Combined MPG']]
```

### MinMax scaler

Hint: Since we are using "numerics" to store the nummerical variables we can pass "numerics" directly
as MinMaxScaler().fit(numerics)

```python
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()

nums_minmaxscaled = pd.DataFrame(minmax.fit_transform(nums), columns = nums.columns)
nums_minmaxscaled.head()
```

### Data Pre-processing (Handling Categorical variables)

```python
cats = vehicles[['Cylinders', 'Fuel Type', 'Drivetrain']].astype('str')
cats.head()
```

### Using One Hot Encoding 

```python
from sklearn.preprocessing import OneHotEncoder

onehot = OneHotEncoder()

cats_onehotcols = []

for i in cats.columns:
    cats_onehotcols = cats_onehotcols + list(i + '-' + np.sort(cats[i].unique()))

cats_onehotenc = pd.DataFrame(onehot.fit_transform(cats).toarray(), columns = cats_onehotcols)
cats_onehotenc.head()
```

```python
X = pd.concat([nums_minmaxscaled, cats_onehotenc], axis=1)
X.head()
```

```python
Y = vehicles[['CO2 Emission Grams/Mile']]
Y.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lm = LinearRegression()
model_sk = lm.fit(X, Y)


print(f'The regression model intercept is {model_sk.intercept_}')
print(f'The regression model coefficients are {model_sk.coef_}, respectively')
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model_sm = ols('Y ~ X', data = vehicles).fit()
model_sm.summary()
```

```python
predictions = model_sk.predict(X)
predictions
```

```python
mse = mean_squared_error(Y, predictions)
rmse = mse**.5
r2 = r2_score(Y, predictions)

print(f'Measures of accuracy:\n')
print(f'MSE = {round(mse, 4)}')
print(f'RMSE = {round(rmse, 4)}')
print(f'R^2 = {round(r2, 4)}')
```
