```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
fig = plt.figure()
```

```python
x1 = np.arange(0, 10, 1)
x1
```

```python
x2 = np.linspace(0, 5, 10)
x2
```

```python
x3 = np.linspace(0, 10, 3)
x3
```

```python
plt.plot(x1, x2, label = 'linear')
plt.plot(x1, x2**2, label = 'quadratic')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Linear vs Non Linear Plots')
plt.legend()

plt.show()
```

```python
x = np.arange(0, 10, .1)

plt.plot(x, np.sin(x), label = 'sine')
plt.plot(x, np.cos(x), label = 'cosine')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Sine vs Cosine Plots')
plt.legend()

plt.show()
```

```python
fig, [ax1, ax2] = plt.subplots(1, 2)
ax1.plot(x1,x2)
ax1.set_title('LINEAR PLOT')   #Note how we are using set_ for subplots
ax1.set_xlabel('x label')      #Note how we are using set_ for subplots
ax1.set_ylabel('y label')      #Note how we are using set_ for subplots

ax2.plot(x, np.cos(x))
ax2.set_title('COSINE PLOT')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')

plt.show()
```

```python
x = np.arange(0, 10, .1)

plt.plot(x, np.sin(x), label = 'sine')
plt.plot(x, np.cos(x), label = 'cosine')
plt.plot(x, np.sin(x + np.pi / 2), label = 'cosine?')

plt.xlabel('x label')
plt.ylabel('y label')
plt.legend()

plt.show()
```

```python
x = np.arange(0, 10, .1)

plt.plot(x, np.sin(x), label = 'sine')
plt.plot(x, np.sin(x + 1 * np.pi / 10))
plt.plot(x, np.sin(x + 2 * np.pi / 10))
plt.plot(x, np.sin(x + 3 * np.pi / 10))
plt.plot(x, np.sin(x + 4 * np.pi / 10))
plt.plot(x, np.sin(x + 5 * np.pi / 10))
plt.plot(x, np.sin(x + 6 * np.pi / 10))
plt.plot(x, np.sin(x + 7 * np.pi / 10))
plt.plot(x, np.sin(x + 8 * np.pi / 10))
plt.plot(x, np.sin(x + 9 * np.pi / 10))
plt.plot(x, np.sin(x + 10 * np.pi / 10))

plt.xlabel('x label')
plt.ylabel('y label')
plt.legend()

plt.show()
```

```python
def my_plots(ax, x, y, params):
    return(ax.plot(x, y, **params))

fig, (ax1, ax2) = plt.subplots(1, 2)
my_plots(ax1, x1, x2, {'marker': 'x'})
my_plots(ax2, x, np.cos(x), {'marker': 'o'})

ax1.set_title('LINEAR PLOT')
ax2.set_title('COSINE PLOT')
```

```python
vehicles = pd.read_csv('vehicles.csv')

vehicles.sort_values(by=["CO2 Emission Grams/Mile"], inplace=True)
vehicles.head()
```

```python
vehicles.plot(x = 'CO2 Emission Grams/Mile', y = ['City MPG', 'Highway MPG', 'Combined MPG'])
```

```python
vehicles.plot(x = 'CO2 Emission Grams/Mile', y = ['City MPG', 'Highway MPG', 'Combined MPG'], figsize = (10,10))
```

```python
vehicle_bar = vehicles[['Highway MPG', 'City MPG', 'Drivetrain']]
vehicles_mean = vehicle_bar.groupby('Drivetrain').agg('mean')
print(vehicles_mean)
vehicles_mean.plot.bar(figsize = (20,10), alpha = 1)
```

```python
vehicle_bar = vehicles[['Highway MPG', 'City MPG', 'Drivetrain']]
vehicles_mean = vehicle_bar.groupby('Drivetrain').agg('mean')
print(vehicles_mean)
vehicles_mean.plot.bar(figsize = (20,10), alpha = .5)
```

```python
vehicles_mean.plot.barh()
```
