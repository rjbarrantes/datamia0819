```python
animals = pd.read_csv('animals.csv')
```

```python
animals.head(3)
```

```python
animals['bodywt10'] = animals['bodywt'] + 10
animals.head(3)
```

```python
animals['body+brain'] = animals['bodywt'] + animals['brainwt']
animals.head(3)
```

```python
animals['wtratio'] = animals['bodywt'] / animals['brainwt']
animals.head(3)
```

```python
animals['wtratiozerocheck'] = np.where(animals['brainwt'] != 0, animals['bodywt'] / animals['brainwt'], 0)
animals.head(3)
```

```python
from scipy import stats

def agg_mode(x):
    return(stats.mode(x)[0])
```

```python
stats.mode(vehicles['Year'])
```

```python
vehicles.groupby(['Transmission'])['Vehicle Class'].agg(agg_mode)
```
