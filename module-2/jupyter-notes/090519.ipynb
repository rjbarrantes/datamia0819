```python
np.random.seed(1000)
seasons = pd.DataFrame(np.random.random((10, 4)), columns = ['winter', 'spring', 'summer', 'autumn'])
lookup = ['summer', 'winter', 'spring', 'summer', 'autumn', 'winter', 'winter', 'spring', 'summer', 'summer'] 
seasons['lookupValues'] = seasons.lookup(seasons.index, lookup)
seasons
```

```python
vehicles = pd.read_csv('vehicles.csv')
```

```python
vehicles.head()
```

```python
vehicles.groupby(['Vehicle Class']).agg({'Cylinders': 'mean'}).head()
```

```python
vehicles.pivot_table(index = ['Vehicle Class'], columns = 'Cylinders').head()
```

```python
vehicles.pivot_table(index = ['Vehicle Class'], values = 'Combined MPG').head()
```

```python
vehicles.pivot_table(index = ['Vehicle Class'], values = 'Combined MPG', aggfunc = np.sum).head()
```

```python
vehicles.pivot_table(index = ['Vehicle Class'], values = 'Combined MPG', aggfunc = np.std).head()
```

```python
vehicles.head(10)
```

```python
plt.scatter(x = 'Highway MPG', y = 'City MPG', data = vehicles)
```

```python
vehicles['City MPG'].corr(vehicles['Highway MPG'])
```

```python
vehicles.corr()
```

```python
correlation = vehicles.corr()
fig, ax = plt.subplots(figsize = (20, 14))
sns.heatmap(correlation, vmax = 1, annot = True, linewidths = 5, ax = ax)
```

```python
plt.scatter(vehicles['City MPG'], vehicles['CO2 Emission Grams/Mile'])
plt.title('Relationship')
plt.xlabel('City MPG')
plt.ylabel('CO2 Emission Grams/Mile')
plt.show()
```

```python
vehicles['City MPG'].corr(vehicles['CO2 Emission Grams/Mile'], method = 'pearson')
```

```python
vehicles['City MPG'].corr(vehicles['CO2 Emission Grams/Mile'], method = 'spearman')
```
