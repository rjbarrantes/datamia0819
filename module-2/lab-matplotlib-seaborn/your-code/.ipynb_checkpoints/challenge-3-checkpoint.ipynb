#### Now in addition to the summary statistics, we also want to visualize the distribution of `Age`. Choose the appropriate plot to sow below.

```python
fig, axes = plt.subplots(1, 2, figsize = (10, 5))

sns.set()
sns.set_style('white')
sns.distplot(titanic['Age'], kde = False, ax = axes[0])
sns.boxplot(x = None, y = 'Age', data = titanic, palette = 'husl', ax = axes[1])

axes[0].set_title('Age Histogram')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')
axes[1].set_title('Age Boxplot')
axes[1].set_ylabel('Age')

fig.subplots_adjust(hspace = 0, wspace = .4)
```

```python
fig = plt.figure(figsize = (10, 10))
grid = plt.GridSpec(3, 1, hspace = .4, wspace = 0)

ax0 = fig.add_subplot(grid[:2, 0])
ax1 = fig.add_subplot(grid[2, 0], sharex = ax0)

sns.set()
sns.set_style('white')
sns.distplot(titanic['Age'], kde = False, ax = ax0)
sns.boxplot(x = 'Age', data = titanic, palette = 'husl', orient = 'h', ax = ax1)

ax0.set_title('Age Histogram')
ax0.set_xlabel('Age')
ax0.set_ylabel('Frequency')
ax1.set_title('Age Boxplot', y = -.12)
ax1.set_xlabel('')
ax1.xaxis.tick_top()
ax1.tick_params(length = 0)
ax1.set_ylabel('')
```

#### What additional information in the previous plot provide you about the feature `Age`? Comment.

```python
# The histogram demonstrates that the distribution is in fact positively skewed, but there also seems to be an unexpected 
# bump at the bottom, showing that there was a sizeable number of infants on-board. The boxplot shows that those just under 
# 60 years of age and older were considered outliers with the respect to the rest of the age distribution.
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the appropriate plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.

*Hint: One of the ways involves using Seaborn*

```python
sns.set()
sns.set_style('white')
sns.relplot(x = 'Fare', y = 'Age', data = titanic)
plt.title('Age by Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
```

```python
sns.set()
sns.set_style('white')
sns.scatterplot(x = 'Fare', y = 'Age', data = titanic)
plt.title('Age by Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
```

```python
sns.set()
sns.set_style('white')
sns.lmplot('Fare', 'Age', data = titanic, fit_reg = True)
```

#### Using Seaborn plot the correlation matrix of various features.

*Hint: search how to use `heatmap`*

```python
mask = np.zeros_like(titanic.corr())
mask[np.triu_indices_from(mask)] = True # removes upper triangle

sns.heatmap(titanic.corr(), mask = mask, cmap = 'Blues', vmin = 0, vmax = 1)
```

```python
fig, axes = plt.subplots(1, 2, figsize = (10, 5))

sns.set()
sns.set_style('white')
sns.distplot(titanic[titanic['Gender'] == 'male']['Age'], kde = False, ax = axes[0])
plt.ylim(0, 150)
plt.xlim(0, 100)
sns.distplot(titanic[titanic['Gender'] == 'female']['Age'], kde = False, ax = axes[1])

axes[0].set_title('Male')
axes[0].set_xlabel('Age')
axes[0].set_xlim(0, 100)
axes[0].set_ylabel('Frequency')
axes[0].set_ylim(0, 150)

axes[1].set_title('Female')
axes[1].set_xlabel('Age')
axes[1].set_xlim(0, 100)
axes[1].set_ylabel('')
axes[1].set_yticks([])
axes[1].set_ylim(0, 150)

fig.subplots_adjust(hspace = 0, wspace = .1)
```
