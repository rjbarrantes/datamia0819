## Coding Session I - 30 minutes

From the Fitbit data, we want to visually understand: 

1. **How the average number of steps change by month. Use the appropriate visualization to show the median steps by month?**
1. **Is Fitbitter more active on weekend or workdays?**

##### All plots must be in the same jupyter notebook cell.

Hints:

- Use `Months_encoded` and `Week_or Weekend` columns
- Use `matplolib.pyplot` object oriented API
- Set your size figure to `12,4`
- Explore `plt.sca`
- Explore `plt.xticks`
- Save your figures

```python
fitbit = pd.read_csv('Fitbit2.csv')
fitbit.loc[fitbit['Work_or_Weekend'] == 0, 'Day Type'] = 'Weekend'
fitbit.loc[fitbit['Work_or_Weekend'] == 1, 'Day Type'] = 'Work'
fitbit.head()
```

```python
fig = plt.figure(figsize=(10, 10))
grid = plt.GridSpec(2, 2, hspace = .3, wspace = .3)

ax1 = fig.add_subplot(grid[0, 0])
ax2 = fig.add_subplot(grid[0, 1])
ax3 = fig.add_subplot(grid[1, :])

sns.barplot(x = 'Months_encoded', y = 'Steps', data = fitbit, palette = 'husl', ax = ax1)
sns.barplot(x = 'Months_encoded', y = 'Steps', data = fitbit, palette = 'husl', ax = ax2, estimator = np.median)
sns.barplot(x = 'Months_encoded', y = 'Steps', data = fitbit, palette = 'Blues', ax = ax3, hue = 'Day Type')

ax1.set_title('Average Steps by Month')
ax2.set_title('Median Steps by Month')
ax2.set_xlabel('Month')
ax2.set_ylabel('')
ax2.set_yticks([])
ax3.set_title('Average Steps by Month and Day Type')
ax3.set_xlabel('Month')
```

```python
fig, axes = plt.subplots(1, 3, figsize = (12, 4))

sns.barplot(x = 'Months_encoded', y = 'Steps', data = fitbit, palette = 'husl', ax = axes[0])
sns.barplot(x = 'Months_encoded', y = 'Steps', data = fitbit, palette = 'husl', ax = axes[1], estimator = np.median)
sns.barplot(x = 'Day Type', y = 'Steps', data = fitbit, palette = 'Blues', ax = axes[2])

axes[0].set_title('Average Steps by Month')
axes[0].set_xlabel('Month')
axes[1].set_title('Median Steps by Month')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('')
axes[2].set_title('Average Steps by Month and Day Type')
axes[2].set_ylabel('')

fig.subplots_adjust(hspace = 0, wspace = .5)
```

## Coding Session II - 30 minutes

**Write a loop to plot 3 scatter plots of the following features:**

1. **`Minutes Lightly Active`  vs `Steps`**
1. **`Minutes Very Active` vs `Steps`**
1. **`Minutes Sedentary` vs `Steps`**

```python
correl_steps = ['Minutes Sedentary', 'Minutes Lightly Active', 'Minutes Fairly Active', 'Minutes Very Active']

fig, axes = plt.subplots(1, len(correl_steps), figsize = (16, 4))

i = 0

for j in correl_steps:
    sns.relplot(x = j, y = 'Steps', data = fitbit, ax = axes[i])
    if i > 0:
        axes[i].set_ylabel('')
        axes[i].set_yticks([])
    i += 1
    plt.close()

fig.subplots_adjust(hspace = 0, wspace = .1)
```

```python
cols = ['Minutes Sedentary', 'Minutes Lightly Active', 'Minutes Fairly Active', 'Minutes Very Active', 'Steps']
df = fitbit[cols]

fig, axes = plt.subplots(1, 4, figsize = (16, 4))

for i in range(4):
    plt.sca(axes[i])
    plt.scatter(df[cols[i]], df['Steps'])
    plt.xlabel(cols[i])
    plt.ylabel('Steps')
```
