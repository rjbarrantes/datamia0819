#### Import all the libraries that are necessary.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0, 100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
fig, [ax1, ax2] = plt.subplots(1, 2)

ax1.plot(x, y, color = 'blue', marker = 'o', linestyle = 'dashed', linewidth = 2, markersize = 0)
ax1.set_title('x-y')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x, z, 'bo--', linewidth = 2, markersize = 0)
ax2.set_title('x-z')
ax2.set_xlabel('x')
ax2.set_ylabel('z')

fig.subplots_adjust(hspace = 0, wspace = .5)

plt.show()
```

```python
def plots(ax, x, y, params):
    return(ax.plot(x, y, **params))

fig, [ax1, ax2] = plt.subplots(1, 2)
plots(ax1, x, y, {'color': 'blue', 'marker': 'o', 'linestyle': 'dashed', 'linewidth': 2, 'markersize': 0})
plots(ax2, x, z, {'color': 'blue', 'marker': 'o', 'linestyle': 'dashed', 'linewidth': 2, 'markersize': 0})

ax1.set_title('x-y')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_title('x-z')
ax2.set_xlabel('x')
ax2.set_ylabel('z')

fig.subplots_adjust(hspace = 0, wspace = .5)
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
x = np.arange(0.01, 10.0, 0.01)

y1 = np.exp(x)
y2 = x**2

fig, ax1 = plt.subplots()
ax1.set_xlabel('x')
ax1.set_ylabel('y = exp(x)', color = 'r')
ax1.plot(x, y1, 'r--')
ax1.tick_params(axis='y')

ax2 = ax1.twinx()
ax2.set_ylabel('y = x^2', color = 'b')
ax2.plot(x, y2, 'b--')
ax2.tick_params(axis = 'y')
plt.show()
```

```python
x = np.arange(0, 100)

plt.plot(x, x**2, label = 'y = x^2')
plt.plot(x, np.exp(x), label = 'y = exp(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic and Exponential Normal')
plt.legend()

plt.xscale('linear')
plt.yscale('linear')

plt.show()
```

```python
x = np.arange(0, 100)

plt.plot(x, x**2, label = 'y = x^2')
plt.plot(x, np.exp(x), label = 'y = exp(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic and Exponential Logarithmic')
plt.legend()

plt.xscale('linear')
plt.yscale('log')

plt.show()
```

#### In the vehicles data set that you have downloaded, use the vehicles.csv file. In this exercise we will conduct some exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. 

Scatter Plot

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
vehicles = pd.read_csv('vehicles.csv')
```

```python
sns.lmplot('Combined MPG', "Highway MPG", data = vehicles)
```

```python
sns.relplot(x = 'Combined MPG', y = 'Highway MPG', data = vehicles)
```

Box Whisker Plot

Please provide a box plot of the variable "CO2 Emission Grams/mile"

```python
sns.boxplot(y = 'CO2 Emission Grams/Mile', data = vehicles)
```

```python
sns.boxplot(x = 'Cylinders', y = 'CO2 Emission Grams/Mile', data = vehicles, palette = 'husl')
```

Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
sns.distplot(vehicles['Fuel Barrels/Year'])
```

```python
sns.distplot(vehicles['Fuel Barrels/Year'], kde=False)
```

Bar Chart
