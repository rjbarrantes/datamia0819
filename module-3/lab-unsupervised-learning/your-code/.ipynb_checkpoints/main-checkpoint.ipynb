```python
def boxplots(df, mult = 1.5):
    import seaborn as sns

    fig = plt.figure(figsize = (20, 11.25))
    sns.set_context('notebook')
    
    return sns.boxplot(data = df, orient = 'h', palette = sns.husl_palette(10, h = .01, s = 1, l = .7), whis = mult)

boxplots(customers.iloc[:, 2:], mult = 2)
```

```python
def outliers_iqr(df, mult = 1.5):
    global outliers
    outliers = []

    for i in df.columns:
        iqri = df.describe()[i]['75%'] - df.describe()[i]['25%']
        loweri = df.describe()[i]['25%'] - mult * iqri
        upperi = df.describe()[i]['75%'] + mult * iqri

        outliers.append(df[(df[i] < loweri) | (df[i] > upperi)])
    outliers = pd.concat(outliers).drop_duplicates().sort_index()
    return outliers

outliers_iqr(customers.iloc[:, 2:])
outliers.shape[0]
```

```python
def outliers_std(df, mult = 2):
    global outliers
    outliers = []

    for i in df.columns:
        stdi = df.describe()[i]['std']
        loweri = df.describe()[i]['mean'] - mult * stdi
        upperi = df.describe()[i]['mean'] + mult * stdi

        outliers.append(df[(df[i] < loweri) | (df[i] > upperi)])
    outliers = pd.concat(outliers).drop_duplicates().sort_index()
    return outliers

outliers_std(customers.iloc[:, 2:])
outliers.shape[0]
```

```python
def missingness(df):
    return df.isnull().sum() / df.shape[0]

missingness(customers)
```

```python
def collinearity(df):
    import seaborn as sns

    fig = plt.figure(figsize = (20, 11.25))
    mask = np.zeros_like(df.corr())
    mask[np.triu_indices_from(mask)] = True
    sns.set_context('notebook')
    
    return sns.heatmap(df.corr(), mask = mask, cmap = 'Blues', annot = True, vmin = 0, vmax = 1)

collinearity(customers.iloc[:, 2:])
```

# Challenge 3 - Data Preprocessing

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). If you made this observation in the first challenge, you've done a great job! This means you not only completed the bonus questions in the previous Supervised Learning lab but also researched deep into [*feature scaling*](https://en.wikipedia.org/wiki/Feature_scaling). Keep on the good work!

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
from sklearn.preprocessing import StandardScaler

customers_scaled = StandardScaler().fit_transform(customers.select_dtypes(exclude = 'object'))
customers_scaled
```

# Challenge 3 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

Count the values in `labels`.

```python
from sklearn.cluster import KMeans

kmeans_fit = KMeans(n_clusters = 3).fit(customers_scaled)
labels_kmeans = pd.Series(kmeans_fit.labels_)

labels_kmeans.value_counts()
```

# Challenge 4 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

Count the values in `labels_DBSCAN`.

```python
from sklearn.cluster import DBSCAN

dbscan_fit = DBSCAN().fit(customers_scaled)
labels_dbscan = pd.Series(dbscan_fit.labels_)

labels_dbscan.value_counts()
```

# Challenge 5 - Compare K-Means with DBSCAN

Now we want to visually compare how K-Means and DBSCAN have clustered our data. We will create scatter plots for several columns. For each of the following column pairs, plot a scatter plot using `labels` and another using `labels_DBSCAN`. Put them side by side to compare. Which clustering algorithm makes better sense?

Columns to visualize:

* `Detergents_Paper` as X and `Milk` as y
* `Grocery` as X and `Fresh` as y
* `Frozen` as X and `Delicassen` as y

```python
from sklearn.decomposition import PCA

pca_customers = pd.DataFrame(PCA().fit_transform(pd.get_dummies(customers)))
pca_customers.columns = pca_customers.columns.astype(str)
pca_customers['labels_kmeans'] = labels_kmeans
pca_customers['labels_dbscan'] = labels_dbscan

fig, ax = plt.subplots(1, 2, figsize = (12, 6))
sns.set_style('whitegrid')
sns.set_context('notebook')

kmeans_palette = sns.husl_palette(len(labels_kmeans.unique()), h = .01, s = 1, l = .7)
dbscan_palette = sns.husl_palette(len(labels_dbscan.unique()), h = .01, s = 1, l = .7)

sns.relplot(x = '0', y = '1', hue = 'labels_kmeans', palette = kmeans_palette, data = pca_customers, ax = ax[0])
plt.close()
sns.relplot(x = '0', y = '1', hue = 'labels_dbscan', palette = dbscan_palette, data = pca_customers, ax = ax[1])
plt.close()
```

# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
customers_kmeans = customers.drop(columns = ['labels_kmeans', 'labels_dbscan'])
customers_kmeans.head()
```

```python
pca_customers_kmeans = pd.DataFrame(PCA().fit_transform((customers_kmeans)))
pca_customers_kmeans.columns = pca_customers_kmeans.columns.astype(str)

fig = plt.figure(figsize = (5, 5))
sns.set_style('whitegrid')
sns.set_context('notebook')

clusters = range(2, 11)
inertia = []

for i in range(len(clusters)):
    kmeans_fit = KMeans(n_clusters = clusters[i]).fit(customers_scaled)
    inertiak = pd.Series(kmeans_fit.inertia_)
    inertia.append(inertiak[0])
inertia

sns.lineplot(x = clusters, y = inertia)
plt.xlabel('k')
plt.ylabel('inertia')
```

```python
pca_customers_kmeans = pd.DataFrame(PCA().fit_transform(customers_kmeans))
pca_customers_kmeans.columns = pca_customers_kmeans.columns.astype(str)

fig, axes = plt.subplots(3, 3, figsize = (15, 15))
sns.set_style('whitegrid')
sns.set_context('notebook')

clusters = range(2, 11)
ax = [(x, y) for x in [0, 1, 2] for y in [0, 1, 2]]

for i in range(len(clusters)):
    
    # kmean labels
    kmeans_fit = KMeans(n_clusters = clusters[i]).fit(customers_scaled)
    labels_kmeans = pd.Series(kmeans_fit.labels_)

    # addition of labels to pca
    pca_customers_kmeans['labels_kmeans'] = labels_kmeans
    kmeans_palette = sns.husl_palette(len(labels_kmeans.unique()), h = .01, s = 1, l = .7)
    
    # scatterplots at various clusters
    sns.relplot(x = '0', y = '1', hue = 'labels_kmeans', palette = kmeans_palette, data = pca_customers_kmeans, 
                ax = axes[ax[i][0], ax[i][1]])
    axes[ax[i][0], ax[i][1]].set_title(f'k = {clusters[i]}')
    plt.close()
```

# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
customers_dbscan = customers.drop(columns = ['labels_kmeans', 'labels_dbscan'])
customers_dbscan.head()
```

```python
pca_customers_dbscan = pd.DataFrame(PCA().fit_transform(customers_dbscan))
pca_customers_dbscan.columns = pca_customers_dbscan.columns.astype(str)

fig, axes = plt.subplots(5, 5, figsize = (25, 25))
sns.set_style('whitegrid')
sns.set_context('notebook')

sizes = [5, 10, 15, 20, 30]
epsilons = [.5, .75, 1, 1.5, 2]
parameters = [(x, y) for x in sizes for y in epsilons]
ax = [(x, y) for x in [0, 1, 2, 3, 4] for y in [0, 1, 2, 3, 4]]

for i in range(len(parameters)):

    # dbscan labels
    dbscan_fit = DBSCAN(eps = parameters[i][1], min_samples = parameters[i][0]).fit(customers_scaled)
    labels_dbscan = pd.Series(dbscan_fit.labels_)

    # addition of labels to pca
    pca_customers_dbscan['labels_dbscan'] = labels_dbscan
    dbscan_palette = sns.husl_palette(len(labels_dbscan.unique()), h = .01, s = 1, l = .7)

    # scatterplots at various clusters
    sns.relplot(x = '0', y = '1', hue = 'labels_dbscan', palette = dbscan_palette, data = pca_customers_dbscan, 
                ax = axes[ax[i][0], ax[i][1]])
    axes[ax[i][0], ax[i][1]].set_title(f'n = {parameters[i][0]}, eps = {parameters[i][1]}')
    plt.close()
```
